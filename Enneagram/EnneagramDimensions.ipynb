```python
import matplotlib.pyplot as plt
import pandas as pd

from collections import defaultdict
from math import pi
```

```python
enneagram_scores = {
    1: 9,
    2: 1,
    3: 8.4,
    4: 2.4,
    5: 9.7,
    6: 8,
    7: 4.7,
    8: 5.4,
    9: 2,
}
```

```python
triads = {
    1: "instictive",
    2: "feeling",
    3: "feeling",
    4: "feeling",
    5: "thinking",
    6: "thinking",
    7: "thinking",
    8: "instictive",
    9: "instictive",
}
hornevian_groups = {
    1: "compliant",
    2: "compliant",
    3: "assertive",
    4: "withdrawn",
    5: "withdrawn",
    6: "compliant",
    7: "assertive",
    8: "assertive",
    9: "withdrawn",
}
harmonic_groups = {
    1: "competency",
    2: "positive_outlook",
    3: "competency",
    4: "reactive",
    5: "competency",
    6: "reactive",
    7: "positive_outlook",
    8: "reactive",
    9: "positive_outlook",
}
```

```python
triad_scores = defaultdict(int)
for num, score in enneagram_scores.items():
    triad_scores[triads[num]] += score

hornevian_scores = defaultdict(int)
for num, score in enneagram_scores.items():
    hornevian_scores[hornevian_groups[num]] += score
    
harmonic_scores = defaultdict(int)
for num, score in enneagram_scores.items():
    harmonic_scores[harmonic_groups[num]] += score
```

```python
triad_scores, hornevian_scores, harmonic_scores
```

```python
def radar_chart_for_dict(dict_):
    categories = list(dict_.keys())
    values = list(dict_.values())
    values = values + [values[0]]
    N = len(categories)

    # What will be the angle of each axis in the plot? (we divide the plot / number of variable)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    # Initialise the spider plot
    ax = plt.subplot(111, polar=True)

    # Draw one axe per variable + add labels
    plt.xticks(angles[:-1], categories, color='white', size=10)

    # Draw ylabels
    ax.set_rlabel_position(0)
    plt.yticks([8,16,24], ["","",""], color="white", size=7)
    plt.ylim(0,24)

    # Plot data
    ax.plot(angles, values, linewidth=1, linestyle='solid')

    # Fill area
    ax.fill(angles, values, 'b', alpha=0.1)

    # Show the graph
    plt.show();
```

```python
plt.figure(figsize=(5,5));
radar_chart_for_dict(triad_scores)
```

```python
plt.figure(figsize=(5,5));
radar_chart_for_dict(hornevian_scores)
```

```python
plt.figure(figsize=(5,5));
radar_chart_for_dict(harmonic_scores)
```
